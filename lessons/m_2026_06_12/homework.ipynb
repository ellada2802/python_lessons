```python
for i in range(3, 10, 2): # от 3 до 9 с шагом 2
    print(i)
```

```python
for i in range(9, 2, -2): # от 9 до 3 с шагом -2
    print(i)
```

```python
fruits = ["яблоко", "банан", "апельсин"]

len_fruits = len(fruits)  # len - длина списка, т.е кол-во элементов в нём
for i in range(len_fruits):  # len_fruits - 3, range(len_fruits) - [0, 1, 2]
    print(i)
```

```python
fruits = ["яблоко", "банан", "апельсин"]

len_fruits = len(fruits)  # len - длина списка, т.е кол-во элементов в нём
for i in range(len_fruits):  # len_fruits - 3, range(len_fruits) - [0, 1, 2]
    print(f"{fruits[i]} - {i}")
```
