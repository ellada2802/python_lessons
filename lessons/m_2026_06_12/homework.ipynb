```python
def range_custom(n):
    l = []
    for i in range(n):
        l.append(i)
    return l

range_custom(10)
```

```python
def range_custom(n_start, n_end, step):
    l = []
    for i in range(n_start, n_end, step):
        l.append(i)
    return l

range_custom(0, 10, 3)
```

### While

```python
"""
Цикл while выполняется пока в условии истина. Цикл while где условие всегла истина True - называется бесконечным
"""

condition = True
while condition:
    pass
```

```python
import time

while True:
    time.sleep(1)
    print(1)
```

```python
"""
Чтобы сдедать цикл while конечным (закончить), то условие должно стать False
"""
```

```python
PASSWORD = "123"
password_from_user = ""

while PASSWORD != password_from_user:
    password_from_user = input("введите корректный пароль")

print("введён корректный пароль")
```

### Изучить и здесь прописать примеры использования команд continue, break (универсальны для обоих циклов)
continue - пропуск шага

break - завершает цикл

```python
PASSWORD = "123"
password_from_user = ""
try_count = 3
count = 0

while PASSWORD != password_from_user:
    password_from_user = input("введите корректный пароль")
    if count > try_count:
        print("кол-во попыток превышено")
        break
    count += 1
```

```python
PASSWORD = "123"
password_from_user = ""
try_count = 3

for i in range(try_count):
    password_from_user = input("введите корректный пароль")
    if PASSWORD == password_from_user:
        print("введён корректный пароль")
        break
else:
    print("кол-во попыток превышено")

print("авторизация закончена")
```
