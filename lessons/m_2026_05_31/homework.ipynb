```python
-- for (что такое цикл for, continue, break)
цикл - команда, которая повторяет блок несколько раз, т.е. не надо постоянно печатать вручную одно и то же
```

```python
fruits = ["яблоко", "банан", "апельсин"]

for fruit in fruits:
    print(f"Я люблю кушать: {fruit}")
```

```python
# ---------- ЗАДАЧА 2: Логин с блокировкой (вложенный if + счётчик попыток) ----------
"""
Условие:
У вас есть правильный логин "user" и пароль "python123".
Пользователь вводит логин и пароль. Программа должна:

1. Если логин неверен → вывести "Логин не найден"
2. Если логин верен, но пароль неверен → вывести "Неверный пароль" и дать ЕЩЁ 2 ПОПЫТКИ 
   ввести пароль (всего 3 попытки)
3. Если пароль верен → вывести "Добро пожаловать!" и завершить программу
4. Если после 3 неудачных попыток пароля → вывести "Доступ заблокирован"

Пример выполнения:
Введите логин: user
Введите пароль: wrong
Неверный пароль. Осталось попыток: 2
Введите пароль: python
Неверный пароль. Осталось попыток: 1
Введите пароль: 123
Неверный пароль. Осталось попыток: 0
Доступ заблокирован
"""

```

```python
LOGIN = "user"
PASSWORD = "python123"

# Попытка 1
login_in = "user"
password_in = "python123"

if login_in == LOGIN and password_in == PASSWORD:
    print("Добро пожаловать")
elif login_in != LOGIN and password_in == PASSWORD:
    print("Логин не найден")
elif login_in == LOGIN and password_in != PASSWORD:
    print("Неверно введен пароль. Осталось попыток 2")

# Попытка 2
login_in = "user"
password_in = "python123"

if login_in == LOGIN and password_in == PASSWORD:
    print("Добро пожаловать")
elif login_in != LOGIN and password_in == PASSWORD:
    print("Логин не найден")
elif login_in == LOGIN and password_in != PASSWORD:
    print("Неверно введен пароль. Осталось попыток 1")

# Попытка 3
login_in = "user"
password_in = "python123"

if login_in == LOGIN and password_in == PASSWORD:
    print("Добро пожаловать")
elif login_in != LOGIN and password_in == PASSWORD:
    print("Логин не найден")
elif login_in == LOGIN and password_in != PASSWORD:
    print("Доступ заблокирован")

```
