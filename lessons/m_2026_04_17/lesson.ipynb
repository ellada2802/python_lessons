#### Числа: целые (int) и дробные (float)

```python
"""
Существуют языки со статической и динамической типизаций.
- Если типизация статическая, то при объявлении переменной в коде мы 
явно указываем тип и обязаны присваивать данной переменной ТОЛЬКО значения 
с данным типом
- Если типизация динамическая (как в Python), то при объявлении переменной в коде мы 
не обязаны явно указывать тип переменной и в данную переменную можно класть и другие типы
"""
```

```python
# int
value: int = 2323232323232323232 # : int - это аннотирование, т.е. мы для  себя указываем тип
print(type(value))
print(getsizeof(value))
```

```python
value = "123" # value - бирка/ссылка, которую можно повесить на ЛЮБОЙ объект
print(value)
print(type(value))
```

```python
from sys import getsizeof

# float
value = 2.2123213213213123  # "=" - присвоение переменно value значения 2.2 (в памяти создаёься объект 2.2)
print(type(value))
print(getsizeof(value))
```

#### bool (булевое значение)

```python
value_true = True
print(value_true)
print(type(value_true))
```

```python
value_false = False
print(value_false)
print(type(value_false))
```

```python
value_bool = 2 == 2  # "==" - оператор стравнения
print(value_bool)
print(type(value_bool))
```

```python
age = 21
if age > 18:
    print("взрослый") # когда "age > 18" истина, т.е. True
else:
    print("ребёнок")  # когда "age > 18" ложь, т.е. False 
```

```python
has_visa = False
if has_visa:
    print("пропускаем")
else:
    print("НЕ пропускам")
```
