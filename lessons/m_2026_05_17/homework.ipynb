```python
# (!) Решение задач можно писать в lessons\m_2026_05_17\homework.ipynb

# ---------- ЗАДАЧА 1: Проверка на диапазон (if-elif-else + граничные условия) ----------
"""
Условие:
Напишите программу, которая запрашивает у пользователя температуру в градусах Цельсия 
и выводит состояние воды:

- "Лёд" — если температура ≤ 0
- "Жидкость" — если температура от 1 до 99 (включительно)
- "Пар" — если температура ≥ 100
- "Неверное значение" — если температура ниже абсолютного нуля (-273)

Пример вывода:
Температура 25 → Жидкость
Температура -5 → Лёд
Температура -300 → Неверное значение
"""
```

```python
# ТВОЙ КОД:
t = int(input("Введите температуру: "))
#t = -10
if t <= 0:
    print("Лёд")
elif t > 1 and t <= 90:
    print("Жидкость")
elif t >= 100:
    print("Пар")
else:
    print("Неверное значение")
```

```python
# ---------- ЗАДАЧА 2: Логин с блокировкой (вложенный if + счётчик попыток) ----------
"""
Условие:
У вас есть правильный логин "user" и пароль "python123".
Пользователь вводит логин и пароль. Программа должна:

1. Если логин неверен → вывести "Логин не найден"
2. Если логин верен, но пароль неверен → вывести "Неверный пароль" и дать ЕЩЁ 2 ПОПЫТКИ 
   ввести пароль (всего 3 попытки)
3. Если пароль верен → вывести "Добро пожаловать!" и завершить программу
4. Если после 3 неудачных попыток пароля → вывести "Доступ заблокирован"

Пример выполнения:
Введите логин: user
Введите пароль: wrong
Неверный пароль. Осталось попыток: 2
Введите пароль: python
Неверный пароль. Осталось попыток: 1
Введите пароль: 123
Неверный пароль. Осталось попыток: 0
Доступ заблокирован
"""

```

```python
LOGIN = "user"
PASSWORD = "python123"

attempts = 3

while attempts >0:
    login_input = input("Введите логин: ")
    password_unput = input("Введите пароль: ")

    #что такое ==? почему input и откуда мне это брать самостоятельно потом? 
    if login_input == LOGIN and password_unput == PASSWORD:
        print("Пароль верный. Доступ разрешен.")
        break
    else:
        attempts = attempts - 1 

    if attempts >0:
        print(f"Пароль неверный. Осталось попыток: {attempts}")
    else: 
        print("Доступ заблокирован")

print("конец")
```
