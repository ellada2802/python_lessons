```python
# (!) Решение задач можно писать в lessons\m_2026_05_17\homework.ipynb

# ---------- ЗАДАЧА 1: Проверка на диапазон (if-elif-else + граничные условия) ----------
"""
Условие:
Напишите программу, которая запрашивает у пользователя температуру в градусах Цельсия 
и выводит состояние воды:

- "Лёд" — если температура ≤ 0
- "Жидкость" — если температура от 1 до 99 (включительно)
- "Пар" — если температура ≥ 100
- "Неверное значение" — если температура ниже абсолютного нуля (-273)

Пример вывода:
Температура 25 → Жидкость
Температура -5 → Лёд
Температура -300 → Неверное значение
"""
```

```python
# ТВОЙ КОД:
t = int(input("Введите температуру: "))
#t = -10
if t <= 0:
    print("Лёд")
elif t > 1 and t <= 90:
    print("Жидкость")
elif t >= 100:
    print("Пар")
else:
    print("Неверное значение")
```
